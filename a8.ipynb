```python
def gradient_descent(x0, y0, f,grad_f, alpha, num_iterations):
    """
    Parameters:
    x0, y0: Initial point for the descent.
    f: a function of two variables
    grad_f: the gradient of f
    alpha: Learning rate.
    num_iterations: Number of iterations to perform.
    Returns:
    """

    x, y = x0, y0 # Initialize x and y with the initial point
    for i in range(num_iterations):
        grad_x, grad_y = grad_f(x,y) 
        # Update x and y by taking a step in the opposite direction of the gradient
        x += -grad_x*alpha # times learning rate, giving the proper step size (LR is adjusted for smaller or larger steps)
        y += -grad_y*alpha
    return x, y
```

```python
def fun_1(x,y):
    return x**2+y**2

def grad_f_1(x,y):
    grad_x = (2*x) # derivative of x**2 is 2x with respect to x, y**2 treated as constant. 
    grad_y = (2*y) 
    return grad_x, grad_y

print(gradient_descent(0.1,0.1,fun_1,grad_f_1,0.1,10))
print(gradient_descent(-1,1,fun_1,grad_f_1,0.01,100))
```

```python
import numpy as np
def fun_2(x,y):
    return 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2) # original function

def grad_f_2(x,y):
    grad_x = 2*x * np.exp(-x**2 - (y-2)**2) + 4*x * np.exp(-x**2 - (y+2)**2) # partial derivative
    grad_y = 2*(y-2) * np.exp(-x**2 - (y-2)**2) + 4*(y+2) * np.exp(-x**2 - (y+2)**2) 
    return grad_x, grad_y

print(gradient_descent(0,1,fun_2,grad_f_2,0.01,10000)) 
print(gradient_descent(0,-1,fun_2,grad_f_2,0.01,10000))
```

```python
import numpy as np
from mpl_toolkits import mplot3d #for 3D plots
import matplotlib.pyplot as plt #usual matplotlib
X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
x,y=np.meshgrid(X,Y)
z= 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2) #function from 4c
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
```
