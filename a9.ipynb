```python
import sympy as sp
x, y = sp.symbols('x y')
```

## (a) 

```python
f = sp.exp(x) * sp.sin(y) + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
df_dx, df_dy
```

## (b) 

```python
g = x**2 * y + x * y**2
gradient=[]
for i in (x,y):
    gradient.append(sp.diff(g,i))

gradatp = []
for expression in gradient:
    gradatp.append(expression.subs({x: 1, y: -1}))

magnitude = sp.sqrt(gradatp[0]**2 + gradatp[1]**2)

print("Gradient of g:", gradient)
print("Gradient at (1, -1):", gradatp)
print("Magnitude of the gradient:", magnitude)
```

## (c) 

```python
h = sp.ln(x**2 + y**2)
hxx = sp.diff(h, x, x)
hyy = sp.diff(h, y, y)
hxy = sp.diff(h, x, y)

print(hxx)
print(hyy)
print(hxy)
```

## (d)

```python
import numpy as np
import matplotlib.pyplot as plt

j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

xs = np.linspace(-3, 3, 400)
ys = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(xs, ys)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title(r'Contour plot')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

## (e) 

```python
m = sp.sin(x + y)
m_func = sp.lambdify((x, y), m, 'numpy')

Z2 = m_func(X, Y)

plt.contourf(X, Y, Z2, levels=50, cmap='inferno')
plt.colorbar()
plt.title(r'Contour plot of $m(x, y) = \sin(x^2 + y^2)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
