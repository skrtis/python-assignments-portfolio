```python
import numpy as np 
import matplotlib.pyplot as plt

#defined the function
def normal_density(mean,variance,x):
    return (1/np.sqrt(2*np.pi*variance))*np.exp(-0.5*((x-mean)**2)/variance)

# plotting a probability distribution with a mean of 0 and a variance of 5
x = np.linspace(-10,10,1000)
y = normal_density(0,5,x)

# plot the graph
plt.plot(x,y)
plt.title('Normal Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

```python
#numerical integration
def integration(mean,variance,a,b):
    # width of a slice
    dx = 1e-5
    
    # number of slices
    n = int((b-a)/dx)

    area = 0
    for i in range(n):
        x = a + i*dx
        y = normal_density(mean,variance,x)
        area = area + y*dx

    return (f'Therefore, the probability of height being in range [{a},{b}] is {area}')
```

```python
# probability that height is in between 162-190 cm
integration(171,(7.1**2),162,190)
```

```python
# the uniform distribution
def uniform_dist(a,b,x):
    if (x >= a and x <= b) or (x >= b and x <= a):
        return 1/(b-a)
    else:
        return 0

def expected_value(prob_function,a,b):
    dx = 1e-5
    slices = int((b-a)/dx)
    expected_val = 0 
    for i in range(slices):
        x = a + i*dx
        y = prob_function(a,b,x)
        expected_val = expected_val + x*y*dx

    return expected_val

# the expected value of the uniform distribution
v = expected_value(uniform_dist,0,1)
print(f'Therefore, the expected value of the uniform distribution is {v}')
```

```python
# exponential distribution
def exponential_dist(rate,x):
    if x >= 0:
      return rate*np.exp(-rate*x)
    else: 
       return 0

def expected_value_for_exp(prob_function,rate,a,b):
    dx = 1e-5
    slices = int((b-a)/dx)
    expected_val = 0 
    for i in range(slices):
        x = a + i*dx
        y = prob_function(rate,x)
        expected_val = expected_val + x*y*dx

    return expected_val


print("Therefore the expected time between pandemics is: "+str(expected_value_for_exp(exponential_dist,1/50,0,50)))

print("This is not what I expected, as I would tend to expect that the expected value would be closer to 50.")
```
